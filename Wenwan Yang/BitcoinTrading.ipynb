**Implement the MACD Trading Strategy**

The idea of MACD(moving average convergence divergence) is long when STMA is higher than LTMA, short when LTMA is higher than STMA.The algorithm here is to make a deision everyday of whether to long or short the bitcoin. At the end of the day, we exit all positions(sell if we longed and buy if we shorted) and make a new trading decision.Now we should compound the returns. The returns should be the cumulative product
$i_{t}=(i_{t-1}+i_{t-1}*r_{t})=(1+r_{t})*i_{t-1}$,$i_{0}=1$

```python
def MACD_Trading_optimized(dataframe):
    result=[]
    for i in range(5,17,2):
        for j in range(25,40,2):
            mom_df=dataframe
            mom_df['short_avg']=pd.rolling_mean(mom_df.weight_price,i)
            mom_df['long_avg']=pd.rolling_mean(mom_df.weight_price,j)
            mom_df['order']=0
            mom_df['order'][mom_df.short_avg>mom_df.long_avg]=1 
            mom_df['order'][mom_df.short_avg<mom_df.long_avg]=-1
            mom_df['returns']=mom_df.weight_price.pct_change()
            #Returns at week t are relative to week t-1. We are buying at week t and seeling at t+1, so we have to adjust by shifting the returns upward.
            mom_df['returns']=mom_df['returns'].shift(-1)
            mom_df['daily_return']=mom_df.order*mom_df.returns
            mom_df['portfolio_value']=(1 + mom_df.daily_return).cumprod()
            result.append((i,j,mom_df['portfolio_value']["2015-03-21"]))
    return result
            
    
```

**Commodity Channel Index Trading Strategy**

CCI meansures a security's variation from the statistical mean.
$\frac{1}{0.015} \frac{p_{t}-SMA(p_{t})}{\sigma(p_{t})}$.
$p_{t}$ is the typical price = $\frac{High+Low+Close}{3}$
When CCI move above +100, generate the buy signal and when CCI moves below -100, generate the sell signal.



```python
def CCI_Trading_Optimized(dataframe):
    result=[]
    for i in range(70,105,5):
        CCI_df=dataframe
        CCI_df['Typical_Price']=(CCI_df.high+CCI_df.low+CCI_df.closed)/3
        CCI_df['SMA_TP']=pd.rolling_mean(CCI_df.Typical_Price,20)
        mean_deviation = lambda x: np.fabs(x - x.mean()).mean()
        CCI_df['mean_dev']=pd.rolling_apply(CCI_df.Typical_Price,20,mean_deviation)
        CCI_df['CCI']=(1/0.015)*(CCI_df.Typical_Price-CCI_df.SMA_TP)/(CCI_df.mean_dev)
        CCI_df['order']=0
        CCI_df['order'][CCI_df.CCI>i]=1 
        CCI_df['order'][CCI_df.CCI<-i]=-1
        CCI_df['returns']=CCI_df.weight_price.pct_change()
        CCI_df['returns']=CCI_df['returns'].shift(-1)
        CCI_df['daily_return']=CCI_df.order*CCI_df.returns
        CCI_df['portfolio_value']=(1 + CCI_df.daily_return).cumprod()
        result.append((i,CCI_df['portfolio_value']["2015-03-21"]))
    return result
```

**Stochastic Oscillator Trading Strategy**

%K = (Current Close - Lowest Low)/(Highest High - Lowest Low) * 100

%D = 3-day SMA of %K

Lowest Low = lowest low for the look-back period

Highest High = highest high for the look-back period

%K is multiplied by 100 to move the decimal point two places

```python
def SO_Trading_Optimized(dataframe):
    result=[]
    for i in range(0,45,5):
        for j in range(60,100,5):
            SO_df=dataframe
            SO_df['previous_closed']=0
            SO_df['previous_closed']=SO_df['closed'].shift(1)
            SO_df['lowest_low']=pd.rolling_min(SO_df.low,14)
            SO_df['highest_high']=pd.rolling_max(SO_df.high,14)
            SO_df['K']=100*(SO_df.previous_closed-SO_df.lowest_low)/(SO_df.highest_high-SO_df.lowest_low)
            SO_df['moving_avg']=pd.rolling_mean(SO_df.K,3)
            SO_df['order']=0
            SO_df['order'][SO_df.K>j]=-1 
            SO_df['order'][SO_df.K<i]=1 
            SO_df['returns']=SO_df.weight_price.pct_change()
            SO_df['returns']=SO_df['returns'].shift(-1)
            SO_df['daily_return']=SO_df.order*SO_df.returns
            SO_df['portfolio_value']=(1 + SO_df.daily_return).cumprod()
            result.append((i,j,CCI_df['portfolio_value']["2015-03-21"]))
    return result
```
